   - values = their marks  

2. Second define a function `print_result(data)` that:  
   - goes through each student (`for name, marks in data.items()`)  
   - applies conditions to assign grade:  
     1. if marks `>= 80` → Grade **A**  
     2. if marks `>= 60` → Grade **B**  
     3. if marks `>= 50` → Grade **C**

# Task 2 — Word Counter
## Description
In this task, we process text data by asking the user to enter a sentence.  
We then split the sentence into words and count how many times each word appears using a dictionary.  
This is a basic skill in **Natural Language Processing (NLP)** where AI models need to analyze and understand text.  

Example:  
If you enter → *"AI is fun and AI is powerful"*  
The output should count → {"AI": 2, "is": 2, "fun": 1, "and": 1, "powerful": 1}  

---

## Steps
1. Take a sentence input from the user.  
2. Split the sentence into words using `split()`.  
3. Create an empty dictionary `counter`.  
4. Loop through each word:  
   - Convert the word into lowercase for uniformity.  
   - If the word exists in dictionary → increase its count.  
   - Else → add it with count 1.  
5. Print the dictionary containing all word counts.  

---

## Practice Variation
Modify the program to also print the **most frequent word** from the dictionary.  

Steps:  
1. Use `max(counter, key=counter.get)` to find the word with the highest frequency.  
2. Use `max(counter.values())` to get its count.  
3. Print both word and count.  

---


```python
sentence=input("Enter a sentence: ")  #Take input sentence from user

words = sentence.split()  #split sentence into list of words

counter = {} #create an empty dictionary for counts

for word in words:  #loop through each word in list
    word = word.lower()  #convert word to lowercase for uniformity
    if word in counter:  #if word already in dictionary, increment its count
        counter[word] += 1
    else:
        counter[word] = 1  #if word not in dictionary, add it with count 1

print("Word counts:" , counter)  #print the word counts


#  Practice variation:
#  Print the most frequent word from the dictionary


sentence=input("Enter a sentence: ")  #Take input sentence from user

words = sentence.split()  #split sentence into list of words

counter = {} #create an empty dictionary for counts

for word in words:  #loop through each word in list
    word = word.lower()  #convert word to lowercase for uniformity
    if word in counter:  #if word already in dictionary, increment its count
        counter[word] += 1
    else:
        counter[word] = 1  #if word not in dictionary, add it with count 1

print("Most frequent word:" , max(counter, key=counter.get), "with count:", max(counter.values()))  #print the most frequent word and its count
```

# Task 6 — Mini Data Analyzer
## Description
In this task, we will **analyze a list of numbers** and calculate:  
- Maximum value  
- Minimum value  
- Average value  

We will also print which numbers are **above the average**.  
This demonstrates how AI (and data science) can **extract insights from raw data**.  

---

## Steps
1. Define a list of numbers (e.g., `[10, 23, 45, 67, 89, 12, 34, 56]`).  
2. Create a function `analyze_data(nums)` that will:  
   - Use `max(nums)` → find maximum.  
   - Use `min(nums)` → find minimum.  
   - Use `sum(nums)/len(nums)` → calculate average.  
3. Print the original list and calculated values.  
4. Find all numbers above the average using a loop or list comprehension.  
5. Print the numbers above the average.  

---

## Practice Variation
- Instead of **hardcoding the list**, take input from the user:  
  - Ask the user to enter numbers separated by spaces.  
  - Use `split()` to divide input into parts.  
  - Convert them into integers using `map(int, ...)`.  
  - Example:  
    - Input: `10 20 30 40`  
    - Output:  
      - Maximum: 40  
      - Minimum: 10  
      - Average: 25  
      - Numbers above average: [30, 40]  

---

## Extra Concept
### `map()` function:
- The `map()` function applies a given function to **each element** of an iterable.  
- Example:  
  - `map(int, ["10", "20", "30"])` → converts all string numbers to integers.  
- Wrapping with `list()` → `list(map(int, ...))` gives a list of integers.  

---


```python
numbers = [10, 23, 45, 67, 89, 12, 34, 56]

def analyze_data(nums):
    maximum = max(nums)  #find maximum value
    minimum = min(nums)  #find minimum value
    average = sum(nums) / len(nums)  #calculate average value

    print("Numbers:", nums) #print the list of numbers
    print("Maximum:", maximum)  #print maximum
    print("Minimum:", minimum)  #print minimum  
    print("Average:", average)  #print average

    # print numbers above average
    above_average = [n for n in nums if n > average]
    print("Numbers above average:", above_average)

analyze_data(numbers)  


# Practice variation:
# Take a list of numbers as input from user instead of hardcoding


numbers = input("Enter numbers separated by spaces: ")  
numbers = list(map(int, numbers.split()))  #convert input string to list of integers
"""Map:
        The map() function applies a given function to all items in an input list (or any iterable) and returns an iterator.(like a list , tuple, etc.) and the map object is then converted to a list using list().
"""
def analyze_data(nums):
    maximum = max(nums)  #find maximum value
    minimum = min(nums)  #find minimum value
    average = sum(nums) / len(nums)  #calculate average value

    print("Numbers:", nums) #print the list of numbers
    print("Maximum:", maximum)  #print maximum
    print("Minimum:", minimum)  #print minimum  
    print("Average:", average)  #print average

    # print numbers above average
    above_average = []
    for n in nums: 
        if n > average:
            above_average.append(n)
    print("Numbers above average:", above_average)  

analyze_data(numbers) 
```

# Wrap-Up Challenge — Student Report System
## Description
This final challenge combines all previous concepts.  
You will build a **student report generator** that:  
- Stores student names and marks in a dictionary.  
- Prints a formatted table showing **Name, Marks, Grade, and Pass/Fail**.  
- Calculates **class average** and identifies the **topper**.  

This simulates how **AI-powered education systems** analyze student performance and generate reports.  

---

## Steps
1. Create a dictionary `students` with keys as student names and values as marks.  
2. Define a function `generate_report(data)` that:  
   - Prints the table header → `Name  Marks  Grade  Result`.  
   - Loops through each student:  
     - Marks `>= 80` → Grade **A**, Result **Pass**  
     - Marks `>= 60` → Grade **B**, Result **Pass**  
     - Marks `>= 50` → Grade **C**, Result **Pass**  
     - Else → Grade **F**, Result **Fail**  
   - Prints each student’s info in a tabular format.  
3. Calculate **class average** → `sum(data.values()) / total_students`.  
4. Identify the **topper** → `max(data, key=data.get)`.  
5. Print class summary: average marks and topper details.  

---

## Practice Variation
- Instead of using a hardcoded dictionary, **accept input** from the user:  
  1. Ask how many students to input (`n`).  
  2. Take each student’s **name** and **marks** as input.  
  3. Build the dictionary dynamically and call `generate_report(students)`.  
  

---


```python
students = {
    "Alice": 85,    
    "Bob": 72,
    "Charlie": 48,
    "David": 90,
    "Eva": 67
}

def generate_report(data):
    print("Name\tMarks\tGrade\tResult")
    print("-"*32)

    for name , marks in data.items():
        if marks >= 80:
            grade , result ="A" , "Pass"
        elif marks >= 60:
            grade , result ="B" , "Pass"
        elif marks >= 50:
            grade , result ="C" , "Pass"
        else:
            grade , result ="F" , "Fail"

        print(f"{name}\t{marks}\t{grade}\t{result}")

    # calculate class average
    average = sum(data.values()) / len(data)

    # find topper
    topper = max(data, key=data.get)

    # print summary
    print("\nClass Average Marks:", round(average,2))
    print("Topper:", topper, "with", data[topper], "marks")

generate_report(students)

# Practice variation:
# Accept % student names and marks as input , then bild and print report

n = 5 #number of students
students = {}
for _ in range(n):
    name = input("Enter student name: ").capitalize()
    while True:
        marks = int(input("Enter marks (0-100): "))
        if marks <0 and marks > 100:
            marks=int(input("Enter marks (0-100): ""))
        else:
            break
                            
    students[name] = marks

def generate_report(data):
    print("Name\tMarks\tGrade\tResult")
    print("-"*32)

    for name , marks in data.items():
        if marks >= 80:
            grade , result ="A" , "Pass"
        elif marks >= 60:
            grade , result ="B" , "Pass"
        elif marks >= 50:
            grade , result ="C" , "Pass"
        else:
            grade , result ="F" , "Fail"

        print(f"{name}\t{marks}\t{grade}\t{result}")

    # calculate class average
    average = sum(data.values()) / len(data)

    # find topper
    topper = max(data, key=data.get)

    # print summary
    print("\nClass Average Marks:", round(average,2))
    print("Topper:", topper, "with", data[topper], "marks")

generate_report(students)
```
