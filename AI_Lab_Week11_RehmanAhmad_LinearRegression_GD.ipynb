# ‚≠ê Mean Squared Error (MSE) ‚Äì Simple Understanding

\[
MSE = \frac{1}{m} \sum_{i=1}^{m} (y_{\text{pred}} - y_{\text{actual}})^2
\]

### üü¢ Simple Explanation
- **y_pred** = the value your model predicts  
- **y_actual** = the real value  
- You subtract them to find the **error**  
- Then **square** the error (so negatives become positive)  
- Do this for **all data points**  
- Add all squared errors  
- Divide by **m** (total number of points) to get the average  

### üü¢ In very easy words:
**MSE tells you how wrong your model is ‚Äî  it finds the average of all squared mistakes.**

# üìö Step 1 ‚Äì Short Theory 

---

## **1.1 Linear Regression (Single Variable)**

We have data in the form:

\[
(x_i, y_i)
\]

Example:  
- **x** = house size  
- **y** = house price  

We want to fit a **straight line** to this data:

≈∑ = wx + b

Where:

- **b (theta zero)** = intercept  
  ‚Üí Predicted value of **y** when **x = 0**

- **w (theta one)** = slope  
  ‚Üí How much **≈∑** changes when **x** increases by 1 unit

### ‚úî Summary  
The goal of Linear Regression is to find the **best-fitting line** through the data.

---

## **1.2 Cost Function (Mean Squared Error)**

We must measure how good or bad our line is.  
For this, we use a formula called the **Cost Function (J)**.

J(w, b) = (1 / 2m) Œ£ (≈∑·µ¢ - y·µ¢)¬≤

Where:

- **m** = number of data points  
- **≈∑·µ¢** = predicted value  
- **y·µ¢** = actual value  

‚úî If predictions are bad ‚Üí **cost increases**  
‚úî If predictions are good ‚Üí **cost decreases**

### ‚úî Summary  
A smaller **J** means a **better** line.

---

## **1.3 Gradient Descent (Simple + Deep Explanation)**

We have a line, but we want the **best possible line**.  
To achieve this, we must **adjust w and b ** again and again.

This adjustment process is done by:

# ‚≠ê Gradient Descent

Gradient Descent basically says:

> ‚ÄúMove step-by-step in the direction where the error decreases.‚Äù

### üîπ Update Rule

w‚±º := w‚±º - Œ± ‚àÇJ/‚àÇw‚±º

Where:

- **Œ∏_j** = parameter (w or b)
- **Œ± (alpha)** = learning rate  
  ‚Üí Size of each step

### üîπ Understanding Learning Rate (Œ±)

| Œ± value | What happens? |
|--------|----------------|
| Too small | Model learns very slowly |
| Too big | Model jumps too far ‚Üí fails (diverges) |
| Just right | Smooth and fast learning |

---

# ‚≠ê Gradient Descent in Ultra Simple Words

Imagine you're standing on a **mountain** (high error).  
You want to go to the **valley** (minimum error).

You cannot see the entire area.  
You only know the **steepest downward direction**.

So you:

1. Look at slopes around you  
2. Take a small step downward  
3. Repeat many times  
4. Eventually reach the lowest point  

This lowest point = **minimum cost**  
At this point, your line (Œ∏ values) is **best**.

---

# ‚≠ê Why Gradient Descent Works?

Because:

- Derivatives tell us the **slope** of the error  
- Slope tells us which direction reduces the error  
- We keep updating parameters until the slope becomes almost zero  

At slope ‚âà 0 ‚Üí  
we reach the **optimal w and b**

---



# üìå Problem Setup

We have training examples (m = 3):

(1,2), (2,2.8), (3,3.6)

We are doing single-variable linear regression:

(y^ = wx + b)

Initial values:

b = 0, w = 0

Learning rate:

Œ± = 0.1

Number of examples:

m = 3

The gradients for Mean Squared Error cost function

(J(w, b) = 1/(2m) Œ£ (y^ - y)^2)

are:

‚àÇJ/‚àÇb = 1/m Œ£ (y^i - yi)

‚àÇJ/‚àÇw = 1/m Œ£ (y^i - yi) xi

Parameter update rule:

Œ∏j := Œ∏j - Œ± (‚àÇJ/‚àÇŒ∏j)

---

# üî¢ Iteration 1 (First Update)

### Step A ‚Äî Compute Predictions

(y^i = w xi + b)

With w = 0 and b = 0:

y^1 = 0, y^2 = 0, y^3 = 0

---

### Step B ‚Äî Compute Errors

error = y^ - y

error1 = 0 - 2 = -2  
error2 = 0 - 2.8 = -2.8  
error3 = 0 - 3.6 = -3.6

Errors = [-2, -2.8, -3.6]

---

### Step C ‚Äî Compute Gradients

‚àÇJ/‚àÇŒ∏0 = (-2 + -2.8 + -3.6)/3 = -2.8

‚àÇJ/‚àÇŒ∏1 = ((-2)*1 + (-2.8)*2 + (-3.6)*3)/3 ‚âà -6.1333

---

### Step D ‚Äî Update Parameters

b := 0 - 0.1 * (-2.8) = 0 + 0.28 = 0.28  
w := 0 - 0.1 * (-6.1333) = 0 + 0.6133 ‚âà 0.6133

After Iteration 1:

b = 0.28, w ‚âà 0.6133

---

# üî¢ Iteration 2 (Second Update)

### Step A ‚Äî Predictions with new thetas

y^i = 0.28 + 0.6133 * xi

y^1 = 0.8933  
y^2 = 1.5066  
y^3 = 2.12

---

### Step B ‚Äî Errors

error1 = 0.8933 - 2 = -1.1067  
error2 = 1.5066 - 2.8 = -1.2933  
error3 = 2.12 - 3.6 = -1.48

Errors ‚âà [-1.1067, -1.2933, -1.48]

---

### Step C ‚Äî Gradients

‚àÇJ/‚àÇb = (-1.1067 + -1.2933 + -1.48)/3 ‚âà -1.2933  
‚àÇJ/‚àÇw = ((-1.1067)*1 + (-1.2933)*2 + (-1.48)*3)/3 ‚âà -2.7111

---

### Step D ‚Äî Update Parameters

b := 0.28 - 0.1 * (-1.2933) ‚âà 0.40933  
Œ∏1 := 0.6133 - 0.1 * (-2.7111) ‚âà 0.8844

After Iteration 2:

Œ∏0 ‚âà 0.4093, Œ∏1 ‚âà 0.8844

---

# üîç Check Cost Function (Optional)

J(Œ∏0, Œ∏1) = 1/(2m) Œ£ (y^ - y)^2

- Initially: J(0,0) = 4.1333  
- After 1st iteration: J(0.28,0.6133) ‚âà 0.84797  
- After 2nd iteration: J(0.4093,0.8844) ‚âà 0.19568

‚úî Cost is decreasing ‚Üí model is improving.

---

# ‚úÖ Summary / Takeaways

- Gradient Descent updates w and b in the direction that reduces cost.  
- Gradients: ‚àÇJ/‚àÇb = average error, ‚àÇJ/‚àÇw = average error weighted by x  
- Learning rate Œ± must be chosen carefully: too small ‚Üí slow learning; too large ‚Üí divergence  
- After multiple iterations, Œ∏ values converge to the best-fitting line.

## 2.2 Implement the Cost Function J(Œ∏‚ÇÄ, Œ∏‚ÇÅ)


````python
# def compute_cost(X , y , thetaw , thetab):
#     m = len(X)
#     y_pred = thetab + thetaw * X
#     cost = (1/(2*m)) * np.sum((y_pred - y) ** 2)
#     return cost

def compute_cost(x , y , w , b):
    m = len(x)
    y_pred = w * x + b
    # print("Y_pred : ", y_pred)
    cost = (1/(2*m)) * np.sum((y_pred - y ) ** 2)
    return cost

print("Cost for w=0, b=0: " , compute_cost(X , y , 0 , 0))
print("Cost for w=1, b=1: " , compute_cost(X , y , 1 , 1))
print("Cost for w=1, b=0.8: " , compute_cost(X , y , 1 , 0.8))
````

### 2.3 Gradient Descent (Single Variable)

````python
def gradient_descent(X  , y , b , w , alpha , epochs):
    m = len(X)
    cost_history = []

    for i in range(epochs):
        y_pred = b + w * X

        d_b = (1/m) * np.sum(y_pred - y)
        d_w = (1/m) * np.sum((y_pred - y) * X)

        # update thetas 
        b = b - alpha * d_b
        w = w - alpha * d_w

        cost = compute_cost(X , y , w , b)
        cost_history.append(cost)

    return b  , w , cost_history

b_init = 0
w_init = 0
alpha = 0.01
epochs = 1000

b_final , w_final , cost_history = gradient_descent(
    X , y , b_init , w_init , alpha , epochs 
)

print("Final b:", b_final)
print("Final w:", w_final)
print("Final cost:", cost_history[-1])
# cost_history
````

### 2.4 Plot Cost vs Iterations

````python
# print(cost_history)
plt.plot(cost_history)
plt.xlabel("Epoch")
plt.ylabel("Cost J(0)")
plt.title("Cost Decrease over Iterations (Single Variable)")
plt.show()
````

### 2.5 Plot Final Regression Line

````python
plt.scatter(X , y , label="Data")
y_hat = b_final + w_final * X
plt.plot(X , y_hat , color="red" , label="Best Fit Line")
plt.xlabel("Study Hours")
plt.ylabel("Marks out of 10")
plt.legend()
plt.title("Linear Regression - Gradient Descent Result")
plt.show()
````

````python
def gradient_descent_multi(X_b, y, theta, alpha, epochs):
    m = len(y)
    cost_history = []

    for i in range(epochs):
        y_pred = X_b.dot(theta)

        # print("y_pred - y : ", y_pred - y)
        gradients = (1 / m) * X_b.T.dot(y_pred - y)
        # print("gradients : ", gradients)
        
        theta = theta - alpha * gradients
        cost_history.append(compute_cost_multi(X_b, y, theta))

    return theta, cost_history

theta_init = np.zeros(X_b.shape[1])
alpha = 0.01
epochs = 1000

theta_final, cost_history_multi = gradient_descent_multi(
    X_b, y, theta_init, alpha, epochs
)

print("Final Œ∏:", theta_final)
print("Final cost:", cost_history_multi[-1])
````

### **Q1: What does the cost function represent?**

**Answer:**
The cost function represents the average error between the predicted values and the actual values in the dataset. It shows how well the model fits the data. A smaller cost value means the predictions are more accurate.

---

### **Q2: How does gradient descent update parameters?**

**Answer:**
Gradient descent updates the model parameters by moving them in the direction that reduces the cost function. It uses the gradient and a learning rate to adjust the parameters step by step until the minimum cost is reached.

---

### **Q3: What is the difference between single-variable and multivariable regression?**

**Answer:**
Single-variable regression uses only one input feature to predict the output. Multivariable regression uses two or more input features. Multivariable regression can model more complex relationships and usually gives better predictions.
