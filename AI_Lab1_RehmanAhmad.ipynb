# Printing Numbers with Python Loops

We will use a `for` loop to print numbers from **1 to 10**.  
Later, we will modify the code to print only **even numbers**.


````python
def factorial(n):
    f = 1
    for val in range(1 , n+1):
        f *= val

    return f


fact=factorial(5)
print("The Factorial of 5 is -> ", fact)

def factorial(n):
    f = 1
    for val in range(1 , n+1):
        f *= val

    return f

x=int(input("Enter Number For Factorial"))
fact=factorial(x)
print(fact)
````
