## Simple Login System (with 3 Attempts)

### Goal: Check username/password with limited tries (3 attempts).

---

### Step 1: Basic Login
1. Define correct credentials:  
   - `curr_user = "admin"`  
   - `curr_pass = "1234"`
2. Set maximum attempts â†’ `attempts = 3`.
3. Use a `while` loop until attempts run out:
   - Ask for `Username` and `Password`.
   - If both match â†’ print `"Login Successfully"` and stop.
   - Else â†’ reduce `attempts` by 1 and show remaining tries.
4. If all attempts finish â†’ print `"Account locked (demo). Please try later."`

---

                 *** Practice Variation ***
## Add a Second Step: Secret PIN Verification

1. After successful login, define an extra variable:  
   - `secret_pin = "7890"`
2. Ask the user for the **Secret PIN**.
3. If entered PIN is correct â†’ print `"Access Granted"`.
4. Else â†’ print `"Wrong PIN! Access Denied"`.




```python
curr_user="admin"
curr_pass="1234"

attemps = 3 

while attemps > 0:
    u = input ("Username: ")
    p = input ("Password: ")

    if u== curr_user and p == curr_pass:
        print("Login Successfully")
        break
    else:
        attemps -=1
        print("Invalid credentials. Attempts left:" , attemps)
if attemps ==0:
    print("Account locked (demo) . Plz try later.")

# Practice variation 
# After a successful login , ask for a secret PIN as a second step.Only then print "Access Granted".

curr_user = "admin"
curr_pass = "1234"
secret_pin = "7890"   # Second step PIN

attempts = 3

while attempts > 0:
    u = input("Username: ")
    p = input("Password: ")

    if u == curr_user and p == curr_pass:
        print("Login Successfully")

        # Step 2: Ask for secret PIN
        pin = input("Enter Secret PIN: ")
        if pin == secret_pin:
            print(" Access Granted")
        else:
            print(" Wrong PIN! Access Denied")
        break   # Exit loop after login attempt
    else:
        attempts -= 1
        print("Invalid credentials. Attempts left:", attempts)

    if attempts == 0:
        print("Account locked (demo). Please try later.")
```

## Add Attempt Limit (5 tries)

1. Add a variable `attempts = 5`.  
2. After each guess, decrease attempts by 1.  
3. If user runs out of attempts:  
   - Print `"Sorry, you've used all your attempts."`  
   - Reveal the secret number.  
4. If guessed correctly before attempts finish, print `"Access Granted"` and stop.


```python
# Task 6 - Number Guessing Game
# Goal : Simple game with loop and conditionals

import random
seceret_number = random.randint(1, 20)  # Random number between 1 and 20
print("I am thinking of a number between 1 and 20.")

while True:
    guess = int(input("Take a guess: "))

    if guess < seceret_number:
        if guess < 0:
            print("Plz Take a guess between 1 and 20.")
        print("Your guess is too low.")
    elif guess > seceret_number:
        if guess > 20:
            print("Plz Take a guess between 1 and 20.")
        print("Your guess is too high.")
    else:
        print("Congratulations! You guessed the number.")
        break  # Exit the loop if guessed correctly

# practice variation:
# limit the number of attempts to 5 .if they fail,reveal the number at the end

import random
seceret_number = random.randint(1, 20)  # Random number between 1 and 20
print("I am thinking of a number between 1 and 20.")
attempts = 5

while attempts > 0:
    guess = int(input("Take a guess: "))

    if guess < seceret_number:
        if guess < 0:
            print("Plz Take a guess between 1 and 20.")
        print("Your guess is too low.")
    elif guess > seceret_number:
        if guess > 20:
            print("Plz Take a guess between 1 and 20.")
        print("Your guess is too high.")
    else:
        print("Congratulations! You guessed the number.")
        break  # Exit the loop if guessed correctly

    attempts -= 1
    print(f"You have {attempts} attempts left.")
    if attempts == 0:
        print(f"Sorry, you've used all your attempts. The number was {seceret_number}.")
```
